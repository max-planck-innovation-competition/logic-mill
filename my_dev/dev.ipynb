```python
import pandas as pd
import requests
from dotenv import dotenv_values

# Load environment variables from .env file
conf = dotenv_values()
API_KEY = conf["API_KEY"]

# API Settings
URL = 'https://api.logic-mill.net/api/v1/graphql/'
headers = {
    'content-type': 'application/json',
    'Authorization': 'Bearer ' + API_KEY,
}
```

```python
# Query all availabe fields in Documents
query_all_fields = """
query IntrospectionQuery {
  __type(name: "Document") {
    name
    fields {
      name
    }
  }
}
"""
response = requests.post(URL, headers=headers, json={'query': query_all_fields})
all_fields = [f.get('name') for f in response.json().get('data').get('__type').get('fields')]
```
